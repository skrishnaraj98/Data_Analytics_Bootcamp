```python
# Create a series from list.
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# Create a series from dictionary: 
data = {"a":1, "b":2, "c":3}
series = pd.Series(data)
print(series)
print(type(series))

# Key thing to note, whatever dict_keys you will provide, it will become the index of the series. 
```

```python
# Create a different custom index. 
data = [10,20,30]
index = ["X","Y","Z"]
series = pd.Series(data,index = index)
print(series)
print(type(series))
```

```python
# Create a DF from a dictionary of list: 

data = {
    "Name":["Krishna", "John", "Michael"],
    "Age": [25,30,34],
    "City": ["Bangalore", "New York", "Mumbai"] 
}

df = pd.DataFrame(data)
print(type(df))
df
```
