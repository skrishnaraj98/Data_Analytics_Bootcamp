```python
## Reading a JSON data: 

import pandas as pd
from io import StringIO
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'

pd.read_json(StringIO(Data))
```

```python
df = pd.read_json(StringIO(Data))
df
```

```python
# Convert df back to json: 

df.to_json()

# Here the problem is that the output json here attaches index with every value unlike the original json.
```

```python
# We can fix this by passing a parameter called "orient" within .to_json() method. 

df.to_json(orient="index")

# This is a kinda by default parameter value, but it still doesnt fix the alteration. 
```

```python
df.to_json(orient="records")

# Yeah the output is same as the original .json data.
```
