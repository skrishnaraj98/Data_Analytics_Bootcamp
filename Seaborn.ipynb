```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn: 
tips_data = sns.load_dataset("tips")            ## .load_dataset is a method which contains online dataset,
                                                # so you just have to select and get started with. 
tips_data
```

```python
## Create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill", y="tip", data = tips_data)
plt.title("Scatter Plot of total bill vs tips")
plt.show()
```

```python
## Line plot

sns.lineplot(x="size", y="total_bill", data = tips_data)
plt.title("Line plot of total bill by size")
```

```python
## Categorical plots: 

## BAr plot: 

sns.barplot(x="day", y="total_bill", data = tips_data)
plt.title("Bar plot of total bill by day")
plt.show()
```

```python
## Box plot: 

sns.boxplot(x="day", y = "total_bill", data = tips_data)
plt.show()
```

```python
## Violin plot: 

sns.violinplot(x="day", y="total_bill", data = tips_data, color= "pink")
```

```python
## Histogram: 

sns.histplot(tips_data["total_bill"], bins= 10, kde= True)
```

```python
## KDE - Kernel Density Estimator: 

sns.kdeplot(tips_data["total_bill"], fill= True)
```

```python
## Pairplot: 

sns.pairplot(tips_data)
```

```python
## Heatmap: 

corr = tips_data[["total_bill", "tip","size"]].corr()

corr
```

```python
sns.heatmap(corr, annot=True, cmap= "coolwarm")
```
