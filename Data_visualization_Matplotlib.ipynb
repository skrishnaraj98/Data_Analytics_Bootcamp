```python
## Histogram: Used to represent the distribution of a dataset. They divide the data into bins and count the number
# of points in each bin. 
```

```python
## Sample data: 

data = [1,1,1,1,2,3,4,5,3,2,3,3,3,1,1,3,5,4,2,2,4]

# Create histogram
plt.hist(data,bins = 5)
```

```python
plt.hist(data,bins = 5, color= "orange",edgecolor="black")
```

```python
# Here in histogram, bin size or width is calculated by dividing Range(Max value - Min value of the data) by 
# Number of Bins (inputted by the user). So in the case above, Max value=5, Min Value = 1, No. of bins = 5
# Bin Width = (5-1)/5 = 4/5 = 0.8
# So bins will be: Starting at 1.0, then 1.0+0.8=1.8, next @ 2.6, 3.4, 4.2 and last 5.0.
```

```python
## Sales Data Visualization: 

# Lets create a sample sales data of 50 rows and 3 columns: 

import pandas as pd
import numpy as np

np.random.seed(40)  # This here is a custom version we setup as version code. This will keep the random numbers
                    # locked in and no matter how many times we execute random module will generate the same numbers.
                    # The value can be any number. 
sales_data = {
    "Product":np.random.choice(["Prod1","Prod2","Prod3"], size= 50),
    "Quantity":np.random.randint(1,100,size=50),
    "Price":np.random.uniform(5,100, size=50)
}

df = pd.DataFrame(sales_data)
df.head(10)
```

```python
#Plot total sales by Products: 

df["Total_Revenue"] = df["Price"]*df["Quantity"]
df["Total_Revenue"]
total_sales_products = df.groupby("Product")["Total_Revenue"].sum()
total_sales_products
```

```python
##Lets create a bar plot:

total_sales_products.plot.bar()
```

```python
total_sales_products.plot(kind = "bar", color = "Teal")
```

```python
## Lets include a date column within the dataset. 

date_display = pd.date_range(start="2022-01-01", periods=50, freq="D")

df["Date"] = date_display
df.head(20)
```

```python
# Lets plot sales over time: 

sales_trend = df.groupby("Date")["Total_Revenue"].sum()

sales_trend

sales_trend.plot(kind="scatter")
```

```python
sales_trend = df.groupby("Date")["Total_Revenue"].sum().reset_index()
sales_trend   ##.reset_index() will turn a series to a dataframe. 
```

```python
plt.plot(sales_trend["Date"],sales_trend["Total_Revenue"])
```
